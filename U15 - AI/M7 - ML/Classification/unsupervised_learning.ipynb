K-means clustering algorithm is one of the well-known algorithms for clustering the data. We need to assume that the numbers of clusters are already known. This is also called flat clustering. It is an iterative clustering algorithm

```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs
```

Step 1: We need to specify the desired number of K subgroups.
Step 2: Fix the number of clusters and randomly assign each data point to a cluster. 
Or in other words we need to classify our data based on the number of clusters.

```python
# generating the two-dimensional dataset, containing four blobs
X, y_true = make_blobs(n_samples=500, centers=4, cluster_std=0.40, random_state=0)
plt.scatter(X[:, 0], X[:, 1], s=50);
plt.show()
```

```python
# initializing K-means for the four clustes
kmeans = KMeans(n_clusters=4)
# train the model
kmeans.fit(X)
# adn predicts
y_kmeans = kmeans.predict(X)
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
# Visualize the machine's findings based on our data, and the fitment according to the number of clusters that are to be found
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```
